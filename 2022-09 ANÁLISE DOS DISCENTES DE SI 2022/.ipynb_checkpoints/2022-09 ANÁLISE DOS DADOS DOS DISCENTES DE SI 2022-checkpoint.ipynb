# ANÁLISE DOS DADOS DOS DISCENTES DE SI 2022

```python
# IMPORTA MÓDULOS

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sn

sn.set(style="whitegrid")
sn.set_palette("RdBu", n_colors=5)

%matplotlib inline

from pylab import rcParams
rcParams['figure.figsize'] = 10, 4
rcParams['font.size'] = 12

sn.color_palette()
```

```python
# IMPORTAR DADOS

dados = pd.read_excel('Curso de Sistemas de Informação (respostas).xlsx')
```

```python
# FUNÇÃO PARA SALVAR OS GRÁFICOS
contador = 1

def salvar_grafico():
    global contador
    nome = str(contador).zfill(5)
    plt.savefig(f'{nome}.png', dpi=600, format='png', bbox_inches='tight')
    
    contador += 1
```

```python
# CRIAR GRÁFICO

fig, ax = plt.subplots()
plt.title('Histrograma da idade de turma', fontsize = 15)
sn.histplot(x = 'Idade', data = dados, kde=True)
plt.xlabel('')
plt.ylabel('')
plt.plot()
salvar_grafico()
```

```python
# CRIAR GRÁFICO

fig, ax = plt.subplots()
plt.title('Distribuição da idade da turma', fontsize = 15)
sn.boxplot(y = 'Idade', data = dados)
plt.xlabel('')
plt.ylabel('')
plt.plot()
salvar_grafico()
```

```python
# CRIAR GRÁFICO

fig, ax = plt.subplots()
plt.title('Histrograma da idade de turma por sexo', fontsize = 15)
sn.histplot(x = 'Idade', data = dados, kde=True, hue='Sexo')
plt.xlabel('')
plt.ylabel('')
plt.plot()
salvar_grafico()
```

```python
# CRIAR GRÁFICO

fig, ax = plt.subplots()
plt.title('Distribuição da idade da turma por sexo', fontsize = 15)
sn.boxplot(y = dados['Idade'],  x = dados['Sexo'])
plt.xlabel('')
plt.ylabel('')
plt.plot()
salvar_grafico()
```

```python
# CRIAR GRÁFICO

fig, ax = plt.subplots()
plt.title('Histograma da idade da turma por estado civil', fontsize = 15)
sn.histplot(x = 'Idade', data = dados, kde=True, hue='Estado Civil')
plt.xlabel('')
plt.ylabel('')
plt.plot()
salvar_grafico()
```

```python
# CRIAR GRÁFICO

fig, ax = plt.subplots()
plt.title('Distribuição da idade da turma por sexo e estado civil', fontsize = 15)
sn.boxplot(y = 'Idade', x = 'Sexo', data = dados, hue = 'Estado Civil')
plt.xlabel('')
plt.ylabel('')
plt.plot()
salvar_grafico()
```

```python
# PADRONIZAR DADOS

for cidadeold, cidadenew in [('Polo em Guanhães', 'Guanhães - MG'), ('Veredinha-MG', 'Veredinha - MG')]:
    dados['Cidade'] = dados['Cidade'].replace(cidadeold, cidadenew)
```

```python
# # CRIAR GRÁFICO

fig, ax = plt.subplots()
plt.title('Quantidade de alunos por cidade', fontsize = 15)
sn.countplot(x = 'Cidade', data = dados)
plt.xlabel('')
plt.ylabel('')
plt.plot()
salvar_grafico()
```

```python
# CRIAR GRÁFICO

fig, ax = plt.subplots()
plt.title('Quantidade de alunos por cidade e sexo', fontsize = 15)
sn.countplot(x = 'Cidade', data = dados, hue = 'Sexo')
plt.xlabel('')
plt.ylabel('')
plt.plot()
salvar_grafico()
```

```python
# CRIAR GRÁFICO

fig, ax = plt.subplots()
plt.title('Quantidade de alunos por cidade e estado civil', fontsize = 15)
sn.countplot(x = 'Cidade', data = dados, hue = 'Estado Civil')
plt.xlabel('')
plt.ylabel('')
plt.plot()
salvar_grafico()
```

```python
# PADRONIZAR DADOS

coluna = 'Você possui uma primeira formação?'

for velho, novo in [('Tecnologia em produção audiovisual',
                                     'SIM - Tecnologia em produção audiovisual'),
                   ('Engenharia Elétrica ', 'SIM - Engenharia Elétrica ')]:
    dados[coluna] = dados[coluna].replace(velho, novo)

formacao, simnao = [], []
for valor in dados[coluna]:
    s_n, f = valor.upper().split(' - ')
    simnao.append(s_n)
    if s_n == 'SIM':
        formacao.append(f)
    else:
        formacao.append('')
        
dados[coluna] = simnao
dados['Formação'] = formacao
```

```python
# CRIAR GRÁFICO

fig, ax = plt.subplots()
plt.title('Quantidade de alunos possuem uma primeira formação', fontsize = 15)
sn.countplot(x = 'Você possui uma primeira formação?', data = dados)
plt.xlabel('')
plt.ylabel('')
plt.plot()
salvar_grafico()
```

```python
# CRIAR GRÁFICO

fig, ax = plt.subplots()
plt.title('Quantidade de alunos possuem uma primeira formação por sexo', fontsize = 15)
sn.countplot(x = 'Você possui uma primeira formação?', data = dados, hue = 'Sexo')
plt.xlabel('')
plt.ylabel('')
plt.plot()
salvar_grafico()
```

```python
# CRIAR GRÁFICO

fig, ax = plt.subplots()
plt.title('Quantidade de alunos possuem uma primeira formação por estado civil', fontsize = 15)
sn.countplot(x = 'Você possui uma primeira formação?', data = dados, hue = 'Estado Civil')
plt.xlabel('')
plt.ylabel('')
plt.plot()
salvar_grafico()
```

```python
# CRIAR GRÁFICO

fig, ax = plt.subplots()
plt.title('Quantidade de alunos por formação', fontsize = 15)
sn.countplot(x = 'Formação', data = dados[dados['Formação'] != ''])
plt.xlabel('')
plt.ylabel('')
plt.xticks(rotation=10)
plt.plot()
salvar_grafico()
```

```python
# CRIAR GRÁFICO

coluna = 'Você trabalha com tecnologia (Desenvolvimento, Rede, Segurança da Informação, Big Data, Outros)?'

dados[coluna] = dados[coluna].apply(lambda x: 'Não Trabalho com tecnologia' if x == 'Não' else 'Só estudo')

fig, ax = plt.subplots()
plt.title('Quantidade de alunos que trabalham com tecnologia', fontsize = 15)
sn.countplot(x = coluna, data = dados)
plt.xlabel('')
plt.ylabel('')
plt.plot()
salvar_grafico()
```

```python
# CRIAR GRÁFICO

coluna = 'Você trabalha com tecnologia (Desenvolvimento, Rede, Segurança da Informação, Big Data, Outros)?'
dados[coluna].value_counts()

fig, ax = plt.subplots()
plt.title('Quantidade de alunos que trabalham com tecnologia por sexo', fontsize = 15)
sn.countplot(x = coluna, data = dados, hue = 'Sexo')
plt.xlabel('')
plt.ylabel('')
plt.plot()
salvar_grafico()
```

```python
# CRIAR GRÁFICO

coluna = 'Você trabalha com tecnologia (Desenvolvimento, Rede, Segurança da Informação, Big Data, Outros)?'
dados[coluna].value_counts()

fig, ax = plt.subplots()
plt.title('Quantidade de alunos que trabalham com tecnologia por estado civil', fontsize = 15)
sn.countplot(x = coluna, data = dados, hue = 'Estado Civil')
plt.xlabel('')
plt.ylabel('')
plt.plot()
salvar_grafico()
```

```python
# PADRONIZAR DADOS

coluna = 'Você usa alguma linguagem de programação no seu dia a dia?'

for valorold, valornew in [('Sim - Python, Sim - R', 'Python e R'), 
                           ('Sim - Java', 'Java'),
                           ('Sim - Python, Sim - Java', 'Java e Python'),
                           ('Não, mas tenho conhecimento de JavaScript e um pouco de Kotltin', 'JavaScript e Kotltin')]:

    dados[coluna] = dados[coluna].replace(valorold, valornew)
```

```python
# CRIAR GRÁFICO

fig, ax = plt.subplots()
plt.title('Quantidade de alunos que programa no dia a dia por linguagem', fontsize = 15)
sn.countplot(x = coluna, data = dados, order = ['Não', 'JavaScript e Kotltin', 'Java', 'Java e Python', 'Python e R'])
plt.xlabel('')
plt.ylabel('')
plt.plot()
salvar_grafico()
```
